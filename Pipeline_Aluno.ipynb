```python
from ml_utils.src import *
import joblib
```

```python
rf_presenca   = joblib.load('models/todos anos/rf_presenca.pkl')
rf_desempenho = joblib.load('models/todos anos/rf_desempenho.joblib')
```

```python
resultados = []

for i in range(100):
    aluno = gerar_aluno_aleatorio()
    resultado = pipeline_aluno(rf_presenca, rf_desempenho, aluno)
    resultados.append(resultado)
```

```python
df = pd.DataFrame(resultados)

chances_df = pd.json_normalize(df['chances_curso'])
df = pd.concat([df.drop(columns='chances_curso'), chances_df], axis=1)
```
