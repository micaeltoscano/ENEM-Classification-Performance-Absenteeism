```python
import pandas as pd
```

```python
colunas_nao_usadas = [
    'NU_INSCRICAO', 'TP_NACIONALIDADE', 'CO_MUNICIPIO_ESC', 'NO_MUNICIPIO_ESC', 'CO_UF_ESC', 'CO_MUNICIPIO_PROVA', 'NO_MUNICIPIO_PROVA', 'CO_UF_PROVA', 
    'CO_PROVA_CN', 'CO_PROVA_CH', 'CO_PROVA_LC', 'CO_PROVA_MT', 'TP_LINGUA', 'TP_ANO_CONCLUIU', 'SG_UF_PROVA', 'SG_UF_ESC'
]

def limpar_dataset(dataset, colunas_nao_usadas):
    df = pd.read_parquet(dataset)
    
    df.drop(columns=colunas_nao_usadas, inplace=True, errors='ignore')

    df = df.dropna()

    return df
```

```python
df = limpar_dataset("enem_parquet/2019", colunas_nao_usadas)

df.to_parquet("dados2019_limpos")
```
