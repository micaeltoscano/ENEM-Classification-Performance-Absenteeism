# Transformação dos Datasets

Os datasets utilizados no projeto do ENEM possuem, em média, 2GB cada, o que implica um custo computacional significativo para armazenamento, leitura e processamento quando utilizados em seu formato original. Diante disso, optamos por converter os dados do formato .csv para Parquet, um formato colunar mais eficiente, que permite melhor compressão dos dados e maior velocidade de leitura, reduzindo assim o consumo de memória e o tempo de processamento durante as análises.

```python
import pandas as pd
import os
```

```python
lista_microdados = ["microdados_enem_2019/DADOS/MICRODADOS_ENEM_2019.csv", "microdados_enem_2020/DADOS/MICRODADOS_ENEM_2020.csv",
                   "microdados_enem_2021/DADOS/MICRODADOS_ENEM_2021.csv", "microdados_enem_2022/DADOS/MICRODADOS_ENEM_2022.csv",
                   "microdados_enem_2023/DADOS/MICRODADOS_ENEM_2023.csv"]

os.makedirs("enem_parquet", exist_ok=True)

for dados in lista_microdados:

    ano = dados[-8:-4]  
    
    pasta_ano = f"enem_parquet/{ano}"
    os.makedirs(pasta_ano, exist_ok=True)

    chunks = pd.read_csv(
        dados,
        sep=";",
        encoding="latin1",
        chunksize=200000
    )
    
    for i, chunk in enumerate(chunks):

        chunk = chunk.convert_dtypes()

        chunk.to_parquet(
            f"{pasta_ano}/part{i}.parquet",
            compression="snappy"
        )
```
